```python
df=pd.read_csv('downloads/mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
df
```

```python
df.head()
```

```python
df['X0'].unique()
```

```python
len(df['X0'].unique())
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

```python
df.X1.value_counts().sort_values(ascending=False).head()
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
lst_10
```

```python
import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
lst_10.append('X1')
df[lst_10]
```
